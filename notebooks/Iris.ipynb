```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.rcParams['figure.figsize'] = (20, 15)

iris = '../data/raw/iris/'
```

```python
df = pd.read_csv(iris+'iris.data', header=None, names=['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'target_names'])
```

```python
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df.target_names = df.target_names.astype('str')
```

```python
for col in df.columns:
    print(len(df[df[col].isna() == True]))
```

There are no nulls in the dataset, now lets see the values if there are any outliers

```python
sns.lineplot(x=df.index, y='sepal_length', data=df, hue='target_names')
plt.title('Iris sepal length');
```

```python
sns.lineplot(x=df.index, y='sepal_width', data=df, hue='target_names')
plt.title('Iris sepal width');
```

```python
sns.lineplot(x=df.index, y='petal_length', data=df, hue='target_names')
plt.title('Iris petal length');
```

```python
sns.lineplot(x=df.index, y='petal_width', data=df, hue='target_names')
plt.title('Iris petal width');
```

We can appreciate that there are no outliers into the data.
